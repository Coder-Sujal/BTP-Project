<big><b>Step 1 — Inspect the CSV</b></big>

```python
import pandas as pd

df = pd.read_csv("../lunar_crater_database_robbins_2018.csv")

print(df.columns)
print(df.head())
```

<big><b>Step 2 — Renaming Columns</b></big>

```python
df = df.rename(columns={
    "LAT_CIRC_IMG": "lat",
    "LON_CIRC_IMG": "lon",
    "DIAM_CIRC_IMG": "diameter"
})
```

<big><b>Step 3 — FILTER YOUR REGION</b></big>

```python
df = df[df["lat"] < -80]
```

```python
df.shape
```

<big><b>Step 4 — Save cleaned CSV</b></big>

```python
df["diameter"] = df["diameter"] * 1000
```

```python
df["lon"] = df["lon"].apply(lambda x: x - 360 if x > 180 else x)
```

```python
df = df[(df["diameter"] > 2000) & (df["diameter"] < 20000)]
```

```python
print("Min diameter:", df["diameter"].min())
print("Max diameter:", df["diameter"].max())
df.shape
```

```python
df.to_csv("../craters_filtered.csv", index=False)
```
